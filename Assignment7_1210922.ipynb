**Select Target and Features use to make the prediction**

```python
# The target we are trying to predict
y = df1['salary']
# The features we will use to make the prediction
X = df1.drop(columns = 'salary')
X
```

**Split data to train and test**

```python
# Import the train and test split from sklearn
from sklearn.model_selection import train_test_split

# Train test split 
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=1210922)
X_train
```

```python
numrical_columns = X_train.select_dtypes("number").columns
ordinal_columns = ['company_size']
nominal_columns = X_train.select_dtypes("object").drop(columns=ordinal_columns).columns
```

**Numrical pipeline**

```python
#import the SimpleImputer ,make_pipeline and StandardScaler from  sklearn
from sklearn.impute import SimpleImputer 
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# instantiate preprocessors
impute_median = SimpleImputer(strategy='median')
scaler = StandardScaler()

# Make a numeric preprocessing pipeline
numrical_pipe = make_pipeline(impute_median, scaler)
numrical_pipe
```

**Ordinal pipline**

```python

impute_na_ordinal = SimpleImputer(strategy='constant', fill_value='NA')

# Specifying the order of categories in quality/condition columns
company_size_order = ["S", "M", "L"]
# Making the list of order lists for OrdinalEncoder
ordinal_category_orders = [company_size_order]
# Instantiate the encoder and include the list of ordered values as an argument
ordinal_encoder = OrdinalEncoder(categories=ordinal_category_orders)

# Making a final scaler to scale category #'s
scaler_ord = StandardScaler()

ordinal_pipe = make_pipeline(impute_na_ordinal, ordinal_encoder, scaler_ord)

ordinal_pipe
```

**Nominal pipline**

```python
impute_na_nominal = SimpleImputer(strategy='constant', fill_value = "NA")

# Instantiate one hot encoder
Nominal_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Make pipeline with imputer and encoder
Nominal_pipe = make_pipeline(impute_na_nominal,Nominal_encoder )
Nominal_pipe
```

**Build pipelines tuples**

```python
from sklearn import set_config
set_config(transform_output='pandas')
numrical_tuple = ('numeric', numrical_pipe, numrical_columns)
ordinal_tuple = ('ordinal', ordinal_pipe, ordinal_columns)
Nominal_tuple = ('categorical', Nominal_pipe, nominal_columns)


# Create ColumnTransformer
# Instantiate with verbose_feature_names_out=False 
col_transformer = ColumnTransformer([numrical_tuple, ordinal_tuple, Nominal_tuple],
                                    verbose_feature_names_out=False)

# Fit on training data
col_transformer.fit(X_train)
# Transform the training data
X_train_transformer = col_transformer.transform(X_train)
# Transform the testing data
X_test_transformer = col_transformer.transform(X_test)
# View the processed training data
X_train_transformer

```

<span style="color:e">**Fit a linear regression model on the transformed data.** </span>

```python
from sklearn.linear_model import LinearRegression
# from sklearn.linear_model SGDRegressor
model = LinearRegression()
model_pipe = make_pipeline(col_transformer, model)
# Fit the model on the training data
model_pipe.fit(X_train, y_train)
model_pipe
```

<span style="color:red">**Print the model parameters and provide an interpretation for each.** </span>

<span style="color:red">**Evaluate the model's quality using appropriate metrics.** </span>

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
def regression_metrics(y_true, y_pred, label='', verbose = True, output_dict=False):
    # Get metrics
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False) 
    r_squared = r2_score(y_true, y_pred)
    if verbose == True:
        # Print Result with Label and Header
        header = "-"*60
        print(header, f"Regression Metrics: {label}", header, sep='\n')
        print(f"- MAE = {mae:,.3f}")
        print(f"- MSE = {mse:,.3f}")
        print(f"- RMSE = {rmse:,.3f}")
        print(f"- R^2 = {r_squared:,.3f}")
    if output_dict == True:
        metrics = {'Label':label, 'MAE':mae,
                 'MSE':mse, 'RMSE':rmse, 'R^2':r_squared}
        
        
# Get predictions for training data
y_train_pred = model_pipe.predict(X_train)

# Call the helper function to obtain regression metrics for training data
results_train = regression_metrics(y_train, y_train_pred, label='Training Data')
print()
# Get predictions for test data
y_test_pred = model_pipe.predict(X_test)
# Call the helper function to obtain regression metrics for test data
results_test = regression_metrics(y_test, y_test_pred, label='Test Data' )
```
